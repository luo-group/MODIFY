# EVmutation local implementation

This code provide guidance on how to re-train EVmutation locally.

```python
from Bio import SeqIO

from evcouplings.utils import read_config_file, write_config_file
from evcouplings.utils.pipeline import execute
```

## Step 1: convert all sequences in MSA to uppercase

```python
msa_location = # path to input MSA file (downloaded from EVcouplings webserver)
msa = SeqIO.parse(msa_location, "fasta")

out_msa = []
for record in msa:
    record.seq = record.seq.upper()
    out_msa.append(record)

out_msa_location = # path to output MSA file
with open(out_msa_location, 'w') as output_handle:
    SeqIO.write(out_msa, output_handle, 'fasta')
```

## Step 2: Revise config file for EVcouplings package

Please change the config file `configs/evmutation_config`:
- Line 28, add path to output prefix (the location that you want to store the EVmutation output)
- Line 30 and 40, add path to output_msa_location in the previous code block
- Line 96-99, add path to executables of plmc and jackhmmer.


## Step 3: Execute EVmutation

After executing this code block, you should be able to see the re-trained EVmutation model under your output prefix.

```python
config_file = '../configs/evmutation_config'

config = read_config_file(config_file)
outcfg = execute(**config)
```
